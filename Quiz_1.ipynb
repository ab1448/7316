The keys are separated from the values by a semi-colon.

````python
people = [
                  {
                      'name': 'Nate',
                      'age': 100,
                      'favorite_color': 'blue',
                      'salary': 250000.19,
                      'mynums': [1, 2, 3, 4, 5]
                  },
                  {
                      'name': 'George',
                      'age': 29,
                      'favorite_color': 'green',
                      'salary': 59000.50
                  },
                  {
                      'name': 'Wendy',
                      'age': 63,
                      'favorite_color': 'red',
                      'salary': 1000000.64
                  },
                  {
                      'name': 'Amy',
                      'age': 47,
                      'favorite_color': 'brown',
                      'salary': 100000.64
                  }
]
````

**5. Using the people list above, print an f-string that displays the name, salary and favorite_color of the 2nd array element like this:**

"```<name>'s salary is <salary> and her favorite color is <favorite_color>.```"

````python
print(f"{people[1]['name']}'s salary is {people[1]['salary']} and his favorite color is {people[1]['favorite_color']}")
````

**6. In the next line, print the same f-string for each person using a for loop.**

**7. Iterate over the ```people``` list and use an if, elif, else statement to print for each person this line:**

```"<name>'s favorite color is a <warm_or_cold_or_other_color> color."```

(Red, orange, and yellow are 'warm' colors; Blue and green are 'cool' colors; All other colors are 'another' color)

````python
for p in people:
  if p['favorite_color'] in ('red', 'orange', 'yellow'):
    print(f"{p['name']}'s favorite color is a warm color")
  elif p['favorite_color'] in ('blue', 'green'):
    print(f"{p['name']}'s favorite color is a cool color")
  else:
    print(f"{p['name']}'s favorite color is another color")
````

**10. Plot the image.**
This is a tricky one because the image (as you see above) is huge. In fact, it is too big to fit in Google Colab's memory. So...read a `rasterio.windows.Window` of the image. See the documentation. I suggest a window not larger than 3000 x 2000. BUT, you can use AI for this question. You can use `matplotlib` or `ee` and `geemap`.

````python
import matplotlib.pyplot as plt
from rasterio.plot import show
import numpy as np

# Compute percentiles
vmin, vmax = np.percentile(d, (2, 98))

window = rasterio.windows.Window(0, 0, 3000, 2000)
d = img.read(window=window)

fig, ax = plt.subplots(figsize=(12, 8))
# show(d, ax=ax)
show(d, ax=ax, vmin=vmin, vmax=vmax)
plt.show()
````
